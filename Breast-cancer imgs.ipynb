```python
classes = ['benign', 'malignent', 'normal']
minority_classes = ['malignent', 'normal']

# Defining minority classes augmentaton t balance em
min_transforms = transforms.Compose([
    RandomHorizontalFlip(p=0.9), #90% prob
    RandomRotation(15, expand=False, center=None),
    ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1), 
])
```

```python
# Data trans
data_transforms = {
    'train': transforms.Compose([
        transforms.Resize(256), transforms.CenterCrop(244),
        #custom trans for min
        transforms.RandomApply([min_transforms], p=0.5) if any(cls in minority_classes for cls in classes) else transforms.RandomApply([], p=0.0),
        transforms.Totensor(), 
    ])
    }
```
